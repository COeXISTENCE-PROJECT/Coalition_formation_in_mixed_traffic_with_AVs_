```python
import numpy as np
import pandas as pd
import os
import pygambit as gbt

def r_matrix(f_path):
    data = []
    for episode_number in range(1, 1025):
        file_path = os.path.join(f_path, f"ep{episode_number}.csv")
        df = pd.read_csv(file_path)
        
        if 'id' in df.columns and 'reward' in df.columns:
            df['episode'] = episode_number
            df = df[df['kind']=='AV']
            data.append(df[['episode', 'id', 'reward']])
    
    full_df = pd.concat(data, ignore_index=True)
    
    reward_matrix = full_df.pivot(index="episode", columns="id", values="reward")
    reward_matrix = reward_matrix.fillna(0)
    
    return reward_matrix
```

```python
f_path_one = r'training_records/all_scenarios_selfish_humans_one/episodes'
f_path_zero = r'training_records/all_scenarios_selfish_humans_zero/episodes'
```

```python
one_r_matrix = r_matrix(f_path_one)
zero_r_matrix = r_matrix(f_path_zero)
```

```python
def convert_to_custom_format(data, title=" ", game_players=" {\"Player 1\" \"Player 2\"}", output_filename="output.txt"):
    """
    Converts a reward matrix (data) into the custom file format and writes it to an output file.
    :param data: 2D list of reward values (e.g., 8x3 matrix).
    :param output_filename: The name of the output file where the result will be saved.
    """
    block_title = title.join(game_players)
    
    block1 = [" { \"1\" \"2\" }" for _ in range(len(data[0]))]  # Assuming "1" and "2" are identifiers for strategies

    # Prepare the second block with rewards
    block2 = []
    for row in data:
        row_str = " ".join(map(str, row))  # Join reward values into a single string
        block2.append(f'{{ "" {row_str} }}')

    # Prepare the last part (the indices of the actions)
    actions = " ".join(map(str, range(1, 1024)))

    # Write everything to the output file
    with open(output_filename, "w") as file:
        file.write(f"{block_title}\n")
        
        file.write("{")
        for line in block1:
            file.write(f"{line}\n")
        file.write{"}\n"}


        file.write('""\n')

        # Write the second block with the rewards
        file.write("{\n")
        for line in block2:
            file.write(f"{line}\n")
        file.write("}\n")

        # Write the actions
        file.write(f"{actions}\n")

    print(f"File has been saved as {output_filename}.")

data = one_r_matrix.to_numpy()
# Call the function to convert and save the file
convert_to_custom_format(data, "zero_equilibrium.txt")
```

```python
one_data = one_r_matrix.to_numpy()
zero_data = zero_r_matrix.to_numpy()

recommended_title = "NFG 1 R \"2^10 two route yield example\"" 
recommended_players = { "Player 1" "Player 2" "Player 3" "Player 4"  "Player 5" "Player 6" "Player 7" "Player 8" "Player 9"  "Player 10"}
recommended_players_string = str(recommended_players)

convert_to_custom_format(one_data, output_filename="one_equilibrium.nfg", title=recommended_title, game_players=recommended_players_string)
convert_to_custom_format(zero_data, output_filename="zero_equilibrium.nfg", title=recommended_title, game_players=recommended_players_string)
```
